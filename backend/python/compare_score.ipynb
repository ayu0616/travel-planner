```python
from test import get_scores
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
scores_list = []
names = [
    "sa",
    "climb_mt",
    # "beam",
    # "define_score",
]
for name in names:
    path = os.path.join("../dist/", name)
    scores_list.append(get_scores(path))
```

```python
li = np.array(scores_list)
li = li.T
df = pd.DataFrame(li, columns=names)
```

```python
df
```

```python
# 折れ線グラフを描画
df.plot()
plt.xlabel("id")
plt.ylabel("score")
plt.show()
```

```python
# ヒストグラムを描画
df.plot.kde()
plt.xlabel("Score")
plt.show()
```

```python
df.plot.box()
plt.ylabel("Score")
plt.show()
```
